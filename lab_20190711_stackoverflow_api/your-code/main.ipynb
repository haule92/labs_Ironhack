# StackAPI

#### Import the necessary libraries here:

```python
# your code here
import json
from stackapi import StackAPI
SITE = StackAPI("stackoverflow")
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
questions = SITE.fetch("questions", fromdate=1559340000, todate=1561931940)
questions
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions = SITE.fetch("questions", fromdate=1559340000, todate=1561931940, tagged="python", sort="votes", min=5)
questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
answers = SITE.fetch("answers", id_s = ["6784","6473"])
answers
```
